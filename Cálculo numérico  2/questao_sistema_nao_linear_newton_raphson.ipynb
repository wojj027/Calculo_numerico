```python
import math
from scipy.misc import derivative
import matplotlib.pyplot as plt
import numpy as np

A = 3.90802e-3
B = 0.580195e-6
R = 200
print(A)

def test_func2(T):
    return 100*(1 + (A*T)- (B*T)**2) - 200

def test_func2_check(T):
    return 100*(1 + (A*T)- (B*T)**2)

def newton_raphson_method(f, x, error_tol, max_iter=16):
    a = f(x) / derivative(f, x)

    iter = 0

    while abs(a) > error_tol and iter < max_iter:
        if iter > max_iter:
            return print(f'Maximum number of {max_iter} iterations exceeded.')

        a = f(x) / derivative(f, x)

        x = x - a
        print(x)
        iter += 1
    return (x, iter)




# x0 = 2
# root, iter = newton_raphson_method(test_func, x0, 0.0001, 16)
root, iter = newton_raphson_method(test_func2, 0, 0.0001, 16)

print(f'Root is approximately {root:.5f} with {iter} iterations')
```

```python
plot_space = np.linspace(0.1, 1000, 10000)  # De 0.1 a 1000 com 1000 amostras
plot_data = [test_func2_check(T) for T in plot_space]

# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(plot_space, plot_data, label=r"$f(Im)$")
plt.xlabel("Temperatura (K)")
plt.ylabel("Resistencia (Ohm)")
plt.xlim(0, 500) # Mostrando somente até 200Hz
plt.legend()
plt.grid(True)
plt.show()
```
